```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = df.fillna(" ")
df['Symptom']=""
for i in range(1,18):
    df['s']=df["Symptom_{}".format(i)]
    df['Symptom']=df['Symptom']+df['s']
```

```python
for i in range(1, 18):
    df = df.drop(f"Symptom_{i}", axis = 1)

df = df.drop('s', axis = 1)
```

```python
class_names = df['Disease'].unique().tolist()
class_names
```

```python
df
```

```python
import numpy as np
plt.figure(figsize=(12, 8))

y_pos = np.arange(len(class_names))

plt.barh(y_pos, [1] * len(class_names), color="skyblue", edgecolor="black")



# Set the labels and title

plt.yticks(y_pos, class_names, fontsize=8)

plt.xlabel('Presence Indicator', fontsize=12)

plt.title('Visualization of Disease Categories', fontsize=14)

plt.tight_layout()

plt.show()
```

```python
df['Disease'] = df['Disease'].apply(lambda x: class_names.index(x))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df['Symptom'], df['Disease'], test_size=0.2)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()

X_train_tfidf = vectorizer.fit_transform(X_train)
```

```python
type(X_train_tfidf)
```

```python


from sklearn.svm import LinearSVC
clf = LinearSVC()
clf.fit(X_train_tfidf,y_train)

```

```python

from sklearn.pipeline import Pipeline
text_clf = Pipeline([('tfidf', TfidfVectorizer()),
                     ('clf', LinearSVC()),])


text_clf.fit(X_train, y_train)  


predictions = text_clf.predict(X_test)

```

```python
predictions
```

```python
from sklearn import metrics
print(metrics.confusion_matrix(y_test,predictions))
```

```python
print(metrics.classification_report(y_test,predictions)) 
```

```python
print(metrics.accuracy_score(y_test,predictions))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
disp = ConfusionMatrixDisplay(confusion_matrix=metrics.confusion_matrix(y_test,predictions),
                              display_labels=text_clf.classes_)
plt.figure(figsize=(30, 6))
disp.plot()
plt.show()
```
